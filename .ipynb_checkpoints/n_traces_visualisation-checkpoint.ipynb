```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_max_corr_n_traces(max_corr_arr, max_true_corr_arr, n_traces_arr, plot_title="", save=False, fig_name="test.svg", fig_folder="./"):   
    plt.ylim(0, 1)
    plt.plot(n_traces_arr, max_corr_arr, linewidth=0.8, color='grey')
    plt.plot(n_traces_arr, max_true_corr_arr, linewidth=0.8, color='red')
    plt.xlabel("Number of Traces")
    plt.ylabel("Highest Correlation Coefficient")
    plt.legend(["Incorrect Weight Values", "True Weight Value"])
    plt.title(plot_title)
    if save:
        plt.savefig(fig_folder + fig_name, format='svg')
    plt.show()
```
