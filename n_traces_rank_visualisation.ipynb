```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_true_weight_rank(rank_arr, n_traces_arr, plot_title="", save=False, fig_name="test.svg", fig_folder="./"):   
    plt.ylim(0, 401)
    plt.plot(n_traces_arr, rank_arr, linewidth=1, color='red')
    #plt.gca().invert_yaxis()
    #plt.gca().set_ylim(bottom=1)
    
    plt.xlabel("Number of Traces")
    plt.ylabel("True Weight Value Rank")
    plt.legend([f'Settled on rank = {rank_arr[-1]}'], loc='upper right')
    plt.title(plot_title)
    if save:
        plt.savefig(fig_folder + fig_name, format='svg')
    plt.show()
```
